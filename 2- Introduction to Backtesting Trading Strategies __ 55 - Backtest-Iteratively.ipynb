```python
import pandas as pd
import yfinance as yf
```

```python
df = yf.download('^GSPC',start='2015-01-01')
```

```python
df
```

```python
df['MA100'] = df.Close.rolling(100).mean()
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df[['Close','MA100']].plot()
```

```python
df['price'] = df.Open.shift(-1)
```

```python
df
```

```python
buyprices,sellprices = [],[]

in_position = False

for index,row in df.iterrows():
    if not in_position and row.Close > row.MA100:
        buyprices.append(row.price)
        in_position = True
    if in_position and row.Close < row.MA100:
        sellprices.append(row.price)
        in_position = False
if in_position:
    sellprices.append(row.Close)
```

```python
len(buyprices),len(sellprices)
```

```python
trades = pd.DataFrame({'buys':buyprices,'sells':sellprices})
```

```python
trades['profit'] = (trades.sells - trades.buys)/trades.buys - 0.005
```

```python
trades
```

```python
(trades.profit + 1).prod() - 1
```
