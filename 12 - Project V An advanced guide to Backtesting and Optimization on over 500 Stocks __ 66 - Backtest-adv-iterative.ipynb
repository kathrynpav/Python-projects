```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
df = yf.download('NVDA',start='2010-01-01')
```

```python
df['price'] = df.Open.shift(-1)
```

```python
def ma_calc(n,m):
    df['sma_1'] = df.Close.rolling(n).mean()
    df['sma_2'] = df.Close.rolling(m).mean()
```

```python
ma_calc(10,20)
```

```python
def backtest(df, n, m):
    ma_calc(n,m)
    in_position = False
    profits = []
    
    for index,row in df.iterrows():
        if not in_position:
            if row.sma_1 > row.sma_2:
                buyprice = row.price
                in_position = True
        if in_position:
            if row.sma_1 < row.sma_2:
                profit = (row.price - buyprice)/buyprice
                profits.append(profit)
                in_position = False
    
    gain = (pd.Series(profits) + 1).prod()
    return gain
```

```python
backtest(df,50,100)
```
